# Simple Regression Practice

```python
# call library
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# generate simple data
dx = np.array([1,2,3,4])
dy = np.array([2,1,4,3])
```

```python
# plot data
plt.scatter(dx, dy)
```

```python
# data process for the regression
dx = dx.reshape(-1,1)
```

## Linear Regression

```python
# call library
from sklearn.linear_model import LinearRegression

# call model
model = LinearRegression()
```

```python
# fit model
model.fit(X = dx, y = dy)
```

```python
# print the results
print("intercepts : ", model.intercept_)
print("coefficient : ", model.coef_)
```

```python
# predction
pred = model.predict(dx)
plt.scatter(dx, dy)
plt.plot(dx, pred)
```

## Multivariate Regression

```python
# data generation
bias = 1.0
beta = np.array([2,3,4,5]).reshape(4,1)
noise = np.random.rand(100,1)
dx = np.random.randn(100,4)
dy = bias + dx.dot(beta)
dy_noise = dy + noise
```

```python
# call model and fit
model = LinearRegression()
model.fit(dx, dy_noise)
```

```python
# print the results
print("intercepts : ", model.intercept_)
print("coefficient : ", model.coef_)
```

```python
# Analytic Solution
bias_dx = np.array([1]*len(dx)).reshape(-1,1)
dx_b = np.hstack([bias_dx, dx])
dxdxt = dx_b.transpose().dot(dx_b)
dxdxt_inv = np.linalg.inv(dxdxt)
beta_analytic = dxdxt_inv.dot(dx_b.transpose()).dot(dy_noise)
print("regression result : ", model.coef_)
print("analytic solution : ", beta_analytic)
```

## Polynomial Regression

```python
# generate data
bias = 1
beta = np.array([2,3]).reshape(2,1)
noise = np.random.randn(100,1)

dx = np.random.randn(100,1)
dx_poly = np.hstack([dx,dx**2])

dy = bias + dx_poly.dot(beta)
dy_noise = dy + noise
plt.scatter(dx, dy_noise)
```

```python
# call model and fit
model = LinearRegression()
model.fit(dx_poly, dy_noise)
```

```python
# print the results
print("intercepts : ", model.intercept_)
print("coefficient : ", model.coef_)
```

```python
# predction
pred = model.predict(dx_poly)
plt.scatter(dx, pred)
```
